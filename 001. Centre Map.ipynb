```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Path to the shapefile
shapefile_path = '/data/IMFSE/PythonCourse/PontdeVilomara/burntArea/burnAreaPdV2022.dbf'

# Load the shapefile
gdf = gpd.read_file(shapefile_path)

# Print the GeoDataFrame and available columns for inspection
print(gdf)
print("Available columns:", gdf.columns)

# Plot the geometries (which represent the burned areas)
gdf.plot()
plt.title('Burned Area Geometries')
plt.show()

# Since no 'burned_area' column exists, we work with the 'geometry' directly
# Ensure that geometries are not empty
if gdf.empty:
    raise ValueError("No geometries found in the shapefile")

# Compute the centroid of the burned area (from the 'geometry' column)
centroid = gdf.geometry.unary_union.centroid

# Plot the geometry with the centroid point marked
gdf.plot()
plt.scatter(centroid.x, centroid.y, color='red', label='Centroid')
plt.legend()
plt.title('Burned Area with Centroid')
plt.show()

# Print centroid coordinates
print(f"Center of burned area: Latitude {centroid.y}, Longitude {centroid.x}")
```
