```python
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# Function to parse the reference time from the time units string
def parse_time_units(units):
    try:
        reference_time_str = units.split('since ')[1]
        reference_time_str = reference_time_str.replace('fire ignition: ', '')
        reference_time = pd.to_datetime(reference_time_str)
    except IndexError:
        raise ValueError(f"Unrecognized time units format: {units}")
    return reference_time

def load_and_process_data_to_df(fileinput, lat, lon, month=7):
    ds = xr.open_dataset(fileinput, decode_times=False)
    print(f"Dataset structure for {fileinput}:")
    print(ds)
    
    if 'time' in ds:
        print(f"Time variable in dataset: {ds['time']}")
        time_units = ds['time'].attrs.get('units', 'seconds since 1970-01-01')
        print(f"Time units: {time_units}")
        reference_time = parse_time_units(time_units)
        time_values = ds['time'].values
        time = [reference_time + timedelta(seconds=float(t)) for t in time_values]
        ds['time'] = ('time', time)
    
    lat_diff = (ds['lat'] - lat)**2
    lon_diff = (ds['lon'] - lon)**2
    dist = np.sqrt(lat_diff + lon_diff)
    y_idx, x_idx = np.unravel_index(dist.argmin(), dist.shape)
    
    if 'time' in ds:
        monthly_data = ds.sel(time=ds.time.dt.month == month)
    else:
        raise ValueError("No valid time dimension found for selecting data by month.")

    data_at_point = monthly_data.isel(x=x_idx, y=y_idx)
    df = data_at_point.to_dataframe().reset_index()

    return df

def load_and_merge_files(file_list, lat, lon, month=7):
    data_frames = []
    
    for file in file_list:
        df = load_and_process_data_to_df(file, lat, lon, month)
        data_frames.append(df)
    
    merged_df = pd.concat(data_frames, ignore_index=True)
    
    return merged_df

def plot_histograms(df, parameters):
    print("Columns in merged DataFrame:")
    print(df.columns)
    
    plt.figure(figsize=(14, 10))
    
    for i, param in enumerate(parameters, 1):
        if param in df.columns:
            plt.subplot(2, 2, i)
            plt.hist(df[param].dropna(), bins=30, color='c', edgecolor='k', alpha=0.7)
            plt.title(f'Histogram of {param}')
            plt.xlabel(param)
            plt.ylabel('Frequency')
        else:
            print(f"Warning: Parameter '{param}' not found in DataFrame.")
    
    plt.tight_layout()
    plt.show()

def calculate_statistics(df, parameters):
    stats = {}
    
    for param in parameters:
        if param in df.columns:
            min_val = df[param].min()
            max_val = df[param].max()
            std_val = df[param].std()
            
            stats[param] = {
                'min': min_val,
                'max': max_val,
                'std': std_val
            }
        else:
            print(f"Warning: Parameter '{param}' not found in DataFrame.")
    
    return pd.DataFrame(stats)

# Example usage
file_list = [
    '/data/IMFSE/PythonCourse/MNH/FCAST_model1.nc',
    '/data/IMFSE/PythonCourse/MNH/FCAST_model2.nc',
    '/data/IMFSE/PythonCourse/MNH/FCAST_model3.nc',
]

lat = 41.709377863541654
lon = 1.892273844304144

merged_data = load_and_merge_files(file_list, lat, lon)

print(merged_data)

merged_data.to_csv('merged_data_mesonh.csv', index=False)

parameters_to_plot = ['si10', 't2m', 'wdir10', 'r2']  # Adjust these based on actual DataFrame columns
plot_histograms(merged_data, parameters_to_plot)

parameters_to_analyze = ['si10', 't2m', 'wdir10', 'r2']  # Adjust these based on actual DataFrame columns
statistics = calculate_statistics(merged_data, parameters_to_analyze)

print(statistics)
```
