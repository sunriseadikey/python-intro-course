```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the MesoNH and CERRA data
mesonh_data = pd.read_csv('mesonh_data.csv')
cerra_data = pd.read_csv('merged_dataCERA.csv')

# Create a figure and set its overall size
plt.figure(figsize=(16, 8))

# Wind Speed Histogram
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st plot
plt.hist(mesonh_data['wspeed'].dropna(), bins=30, color='#1f77b4', alpha=0.7, label='MesoNH Wind Speed', edgecolor='black')
plt.hist(cerra_data['si10'].dropna(), bins=30, color='#ff7f0e', alpha=0.7, label='CERRA Wind Speed', edgecolor='black')
plt.title('Wind Speed Comparison', fontsize=16, fontweight='bold')
plt.xlabel('Wind Speed (m/s)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)

# Wind Direction Histogram
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd plot
plt.hist(mesonh_data['wdir'].dropna(), bins=30, color='#1f77b4', alpha=0.7, label='MesoNH Wind Direction', edgecolor='black')
plt.hist(cerra_data['wdir10'].dropna(), bins=30, color='#ff7f0e', alpha=0.7, label='CERRA Wind Direction', edgecolor='black')
plt.title('Wind Direction Comparison', fontsize=16, fontweight='bold')
plt.xlabel('Wind Direction (degrees)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)

# Set a tight layout to avoid overlap and improve spacing
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the MesoNH and CERRA data
mesonh_data = pd.read_csv('mesonh_data.csv')
cerra_data = pd.read_csv('merged_dataCERA.csv')

# Create a figure and set its overall size
plt.figure(figsize=(16, 8))

# Wind Speed Histogram (Normalized with PDF)
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st plot
plt.hist(mesonh_data['wspeed'].dropna(), bins=30, color='#1f77b4', alpha=0.7, label='MesoNH Wind Speed', edgecolor='black', density=True)
plt.hist(cerra_data['si10'].dropna(), bins=30, color='#ff7f0e', alpha=0.7, label='CERRA Wind Speed', edgecolor='black', density=True)
plt.title('Wind Speed Comparison (Normalized by PDF)', fontsize=16, fontweight='bold')
plt.xlabel('Wind Speed (m/s)', fontsize=14)
plt.ylabel('Probability Density', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)

# Wind Direction Histogram (Normalized with PDF)
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd plot
plt.hist(mesonh_data['wdir'].dropna(), bins=30, color='#1f77b4', alpha=0.7, label='MesoNH Wind Direction', edgecolor='black', density=True)
plt.hist(cerra_data['wdir10'].dropna(), bins=30, color='#ff7f0e', alpha=0.7, label='CERRA Wind Direction', edgecolor='black', density=True)
plt.title('Wind Direction Comparison (Normalized by PDF)', fontsize=16, fontweight='bold')
plt.xlabel('Wind Direction (degrees)', fontsize=14)
plt.ylabel('Probability Density', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)

# Set a tight layout to avoid overlap and improve spacing
plt.tight_layout()

# Display the plot
plt.show()
```
