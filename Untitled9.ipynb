```python
import geopandas as gpd
import matplotlib.pyplot as plt
import xarray as xr
import numpy as np
from shapely.geometry import Point

# Path to your dataset
fileinput = '/data/IMFSE/PythonCourse/CDS/CERRA/cerra-2012.nc'

# Coordinates you are interested in
lat = 41.709377863541654
lon = 1.892273844304144

# Load dataset
ds = xr.open_dataset(fileinput)

# Calculate the nearest grid point indices
abs_diff_lat = np.abs(ds.latitude[:, 0].values - lat)
abs_diff_lon = np.abs(ds.longitude[0, :].values - lon)
lat_idx = abs_diff_lat.argmin().item()
lon_idx = abs_diff_lon.argmin().item()

# Get the actual lat/lon of the nearest grid point (center of interest)
nearest_lat = ds.latitude[lat_idx, 0].values
nearest_lon = ds.longitude[0, lon_idx].values

# Create a GeoDataFrame with the centroid (in this case, it's just the nearest point)
centroid = Point(nearest_lon, nearest_lat)
gdf = gpd.GeoDataFrame(geometry=[centroid])

# Plot the point
gdf.plot()
plt.scatter(nearest_lon, nearest_lat, color='red', label='Centroid')
plt.legend()
plt.title(f'Center of the Grid Point: Latitude {nearest_lat}, Longitude {nearest_lon}')
plt.show()

# Print the center of the grid point
print(f"Center of grid point: Latitude {nearest_lat}, Longitude {nearest_lon}")
```
